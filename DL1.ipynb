```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('/content/rgv.jpg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(img)
```

```python
plt.imshow(gray_img)
```

```python
hist1 =cv2.calcHist([gray_img],[0],None,[256],[0,256])
img_2 =cv2.equalizeHist(gray_img)
hist2 =cv2.calcHist([img_2],[0],None,[256],[0,256])
plt.subplot(221),plt.imshow(gray_img);
plt.subplot(222),plt.plot(hist1);
plt.subplot(223),plt.imshow(img_2);
plt.subplot(224),plt.plot(hist2);
```

```python
img = cv2.imread('/content/badapple.jpg')
img2=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
gray_img1 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img1)
```

```python
th, th_img = cv2.threshold(gray_img1,200,155,cv2.THRESH_BINARY)
plt.imshow(th_img,cmap="gray")
```

```python
th, th_img = cv2.threshold(gray_img1,200,155,cv2.THRESH_BINARY_INV)
plt.imshow(th_img,cmap="gray")
```

```python
plt.imshow(np.flipud(gray_img1),cmap='gray')
```

```python
plt.imshow(np.fliplr(gray_img1),cmap='gray')
```

```python
img = cv2.resize(img,(500,500),interpolation = cv2.INTER_AREA)
gry = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
blr = cv2.GaussianBlur(gry,(5,5),0)
edg = cv2.Canny(blr,80,120)
```

```python
plt.imshow(edg, cmap="gray")
```

```python
ero = cv2.erode(edg,(11,11),iterations=1)
plt.imshow(ero)
```

```python
dila = cv2.dilate(edg,(7,7), iterations=10)
plt.imshow(dila)
```
